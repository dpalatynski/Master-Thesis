```python
import random
import networkx as nx
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
G = nx.read_gpickle('network_neighbourhood.pickle')
```

```python
with open('neighbourhood_position.pickle', 'rb') as handle:
    pos = pickle.load(handle)
```

```python
L = nx.star_graph(len(G))
pos[66] = (-0.1, -0.1)
mapping = {0:66, 66:0}
L = nx.relabel_nodes(L, mapping)

attrs = {(66, i): {"media": random.choices([-1, 1], [0.9, 0.1])} for i in range(len(G.nodes()))}
edge_colors = ['red' if attrs[(66, i)]['media'][0] == 1 else 'grey' for i in range(len(attrs))]
node_colors = ['black'] + ['red' if attrs[(66, i)]['media'][0] == 1 else 'grey' for i in range(len(attrs))]
nx.set_edge_attributes(L, attrs)

plt.figure(figsize=(16, 9))
plt.title('Media influence network on neighbourhood network', fontsize=20)
nx.draw(L, edge_color=edge_colors, node_color=node_colors, pos=pos, with_labels=True, node_size=[800]*len(L))
```
